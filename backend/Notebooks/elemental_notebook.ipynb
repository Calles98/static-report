```python
import pandas as pd 
import numpy as np 
import plotly.graph_objects as go
import random
```

```python
data_filepath = 'FILE_PATH_PLACEHOLDER'
elemental_data = pd.read_csv(data_filepath)
elemental_data = elemental_data.rename(columns={"0": "Sample"})
elemental_data = elemental_data.set_index("Sample")
```

```python
fig = go.Figure()
for sample in elemental_data.index:
  fig.add_trace(go.Bar(x=elemental_data.columns, y=elemental_data.loc[sample], name=sample))

limit_lines_nom147 = {
    "Ag": 100,
    "As": 100,
    "Pb": 100,
    "Sb": 10.6,
    "Cr": 100,
    "Ba": 2000,
    "Se": 20,
    "Be": 24.4, 
    "Hg": 4, 
    "Cd": 20,
  
}

limit_lines_nom157 = {
    "Ni": 1600,
    "Tl": 5.2,
    "V": 78, 
}

# Add limit lines for each element
for i, col in enumerate(elemental_data.columns):
    if col in limit_lines_nom147:
        fig.add_shape(
            type="line",
            x0=i - 0.4,  # start before the box
            x1=i + 0.4,  # end after the box
            y0=limit_lines_nom147[col],
            y1=limit_lines_nom147[col],
            line=dict(color="black", width=2, dash="solid"),
            xref="x", yref="y"
        )
    if col in limit_lines_nom157:
        fig.add_shape(
            type="line",
            x0=i - 0.4,  # start before the box
            x1=i + 0.4,  # end after the box
            y0=limit_lines_nom157[col],
            y1=limit_lines_nom157[col],
            line=dict(color="black", width=2, dash="dot"),
            xref="x", yref="y"
        )
      

# --- Lock y-axis range ---
fig.update_layout(
    yaxis=dict(
        type="log",
        #range=[np.log10(ymin), np.log10(ymax)],  # <- prevents rescaling
        title="Concentración (ppm)"
    ),
    xaxis_title="Elemento",
    barmode="group",
    title="Resultados de Análisis Elemental de muestras de Tepetate - Pila y Antigua Tepetatera",
    legend_title="Muestra",
    template="plotly_white",
    height=700,
    width=1400
)

fig.show()
```
