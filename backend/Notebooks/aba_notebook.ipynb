```python
import pandas as pd 
import numpy as np 
import plotly.graph_objects as go 
import random
#from IPython.display import display
import warnings
warnings.filterwarnings('ignore')
```

```python
file_path = FILEPATH_PLACEHOLDER
display(file_path)
data = pd.read_csv(file_path)
```

<h1 align="center">ABA</h1>

```python
data.loc[data['Descripción'].str.contains('Antiguo'), 'Descripción'] = 'Tepetate Antiguo'
```

```python
types = ['Pila de Tepetate', 'Tepetate Antiguo', 'Tepetate - Lutita',
       'Tepetate - Toba', 'Tepetate - Metamorfizado',
       'Tepetate - Sulfuro Masivo', 'Tepetate - Esquisto', 'Presa de Jales 1-M1 (2020)', 'Mineral', 'Material de Préstamo']

data['Group'] = (data['Descripción'] != data['Descripción'].shift()).cumsum()
groups = {key: group.drop(columns='Group') for key, group in data.groupby('Group')}
```

```python
marker_symbols = ["triangle-up", "square", "x", "cross", "circle", "diamond"]
fig = go.Figure()
for group in groups:
  if any(type in groups[group].iat[0,1] for type in types):
    #print(groups[group].iat[0,1])
    fig.add_trace(go.Scatter(x=groups[group]['Potencial de Neutralización de Ácido (NP)'], y=groups[group]["Potencial de Generación de Ácido (AP)"], mode="markers", marker_symbol=random.choice(marker_symbols), name=groups[group].iat[0,1]))
fig.add_trace(go.Scatter(x=[0, 350], y=[0, 350/1.2], mode="lines", name="1.2 NPR"))
fig.add_trace(go.Scatter(x=[0, 350], y=[0, 350/3], mode="lines", name="3 NPR"))


fig.add_annotation(
    x=300,
    y=350/1.2,
    text="Potencialmente<br> Generador de Drenaje Ácido",
    showarrow=False,
    bordercolor="black",
    borderwidth=1,
    bgcolor="white",
    font=dict(size=12),
    #textangle=-10,
)
fig.add_annotation(
    x=300,
    y=(350/3) + 10,
    text="Incierto/Potencialmente Peligroso",
    showarrow=False,
    bordercolor="black",
    borderwidth=1,
    bgcolor="white",
    font=dict(size=12),
    #textangle=-10
)

fig.add_annotation(
    x=300,
    y=(350/(3*5))-10,
    text="No Potencialmente<br> Generador de Drenaje Ácido",
    showarrow=False,
    bordercolor="black",
    borderwidth=1,
    bgcolor="white",
    font=dict(size=12),
    #textangle=-10
)



fig.update_layout(title='Scatter Plot', xaxis_title='Potencial de Neutralización de Ácido "NP" (t/kt)', yaxis_title='Potencial de Generación de Ácido "AP" (t/kt)')
```

```python
fig = go.Figure()
for group in groups:
    if any(type in groups[group].iat[0, 1] for type in types):
        fig.add_trace(go.Scatter(
            x=groups[group]['NNP B'],
            y=groups[group]['NAGpH A'],
            mode="markers",
            marker_symbol=random.choice(marker_symbols),
            name=groups[group].iat[0, 1]
        ))

# Reference limit lines
fig.add_trace(go.Scatter(x=[20, 20], y=[0, 9], mode="lines", name="NNP = 20", line=dict(color="black", dash="dash")))
fig.add_trace(go.Scatter(x=[-20, -20], y=[0, 9], mode="lines", name="NNP = -20", line=dict(color="black", dash="dash")))
fig.add_trace(go.Scatter(x=[-300, 250], y=[4.5, 4.5], mode="lines", name="NAGpH = 4.5",line=dict(color="black", dash="dash")))

# Add text labels for limit lines
fig.add_annotation(x=20-4, y=9-2, text="NNP = 20", showarrow=False, textangle=-90,font=dict(size=12, color="black", weight="bold"))
fig.add_annotation(x=-20-4, y=9-2, text="NNP = -20", showarrow=False, textangle=-90,font=dict(size=12, color="black", weight="bold"))
fig.add_annotation(x=-250-3, y=5, text="NAGpH = 4.5", showarrow=False,font=dict(size=12, color="black", weight="bold"))


# Add shaded regions to indicate sub-areas
fig.add_shape(type="rect", x0=-300, x1=-20, y0=0, y1=4.5,
              fillcolor="red", opacity=0.2, line_width=0,
              name="PAF")  # Potentially Acid Forming

fig.add_shape(type="rect", x0=-20, x1=20, y0=0, y1=9,
              fillcolor="yellow", opacity=0.2, line_width=0,
              name="Uncertain")

#fig.add_shape(type="rect", x0=20, x1=250, y0=1, y1=4.5,
 #             fillcolor="yellow", opacity=0.2, line_width=0,
  #            name="Low Risk")

fig.add_shape(type="rect", x0=-300, x1=-20, y0=4.5, y1=9,
              fillcolor="orange", opacity=0.2, line_width=0,
              name="Non-Acid Forming")

fig.add_shape(type="rect", x0=20, x1=250, y0=4.5, y1=9,
              fillcolor="green", opacity=0.2, line_width=0,
              name="Non-Acid Forming")

# Add text labels for these sub-areas
fig.add_annotation(x=-150, y=2.5, text="Potencial Generador de Drenaje de Ácido", showarrow=False, font=dict(size=14, color="black"))
fig.add_annotation(x=0, y=3.5, text="Incierto/Potencialmente peligroso", showarrow=False, textangle=-90,font=dict(size=14, color="black"))
fig.add_annotation(x=150, y=7, text="Sin Potencial Generación de Drenaje Ácido", showarrow=False, font=dict(size=14, color="black"))
fig.add_annotation(x=-150, y=7, text="Posible Potencial Generador de Drenaje Ácido", showarrow=False, font=dict(size=14, color="black"))

# Layout
fig.update_layout(
    title='NNP vs NAGpH Classification',
    xaxis_title='Potencial Neto de Neutralización (NNP, t/kt)',
    yaxis_title='NAGpH (u.s)',
    plot_bgcolor="white"
)


```
