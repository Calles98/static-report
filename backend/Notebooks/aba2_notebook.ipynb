```python
import pandas as pd 
import numpy as np 
import plotly.graph_objects as go 
import random
#from IPython.display import display
import warnings
warnings.filterwarnings('ignore')
```

```python
file_path = FILEPATH_PLACEHOLDER
display(file_path)
data = pd.read_csv(file_path)
```

<h1 align="center">ABA</h1>

```python
data.loc[data['Descripción'].str.contains('Antiguo'), 'Descripción'] = 'Tepetate Antiguo'
```

```python
types = ['Pila de Tepetate', 'Tepetate Antiguo', 'Tepetate - Lutita',
       'Tepetate - Toba', 'Tepetate - Metamorfizado',
       'Tepetate - Sulfuro Masivo', 'Tepetate - Esquisto', 'Presa de Jales 1-M1 (2020)', 'Mineral', 'Material de Préstamo']

data['Group'] = (data['Descripción'] != data['Descripción'].shift()).cumsum()
groups = {key: group.drop(columns='Group') for key, group in data.groupby('Group')}
```

```python
marker_symbols = ["triangle-up", "square", "x", "cross", "circle", "diamond"]
fig = go.Figure()
for group in groups:
  if any(type in groups[group].iat[0,1] for type in types):
    #print(groups[group].iat[0,1])
    fig.add_trace(go.Scatter(x=groups[group]['Potencial de Neutralización de Ácido (NP)'], y=groups[group]["Potencial de Generación de Ácido (AP)"], mode="markers", marker_symbol=random.choice(marker_symbols), name=groups[group].iat[0,1]))
fig.add_trace(go.Scatter(x=[0, 350], y=[0, 350/1.2], mode="lines", name="1.2 NPR"))
fig.add_trace(go.Scatter(x=[0, 350], y=[0, 350/3], mode="lines", name="3 NPR"))


fig.add_annotation(
    x=300,
    y=350/1.2,
    text="Potencialmente<br> Generador de Drenaje Ácido",
    showarrow=False,
    bordercolor="black",
    borderwidth=1,
    bgcolor="white",
    font=dict(size=12),
    #textangle=-10,
)
fig.add_annotation(
    x=300,
    y=(350/3) + 10,
    text="Incierto/Potencialmente Peligroso",
    showarrow=False,
    bordercolor="black",
    borderwidth=1,
    bgcolor="white",
    font=dict(size=12),
    #textangle=-10
)

fig.add_annotation(
    x=300,
    y=(350/(3*5))-10,
    text="No Potencialmente<br> Generador de Drenaje Ácido",
    showarrow=False,
    bordercolor="black",
    borderwidth=1,
    bgcolor="white",
    font=dict(size=12),
    #textangle=-10
)



fig.update_layout(title='Scatter Plot', xaxis_title='Potencial de Neutralización de Ácido "NP" (t/kt)', yaxis_title='Potencial de Generación de Ácido "AP" (t/kt)')
```
